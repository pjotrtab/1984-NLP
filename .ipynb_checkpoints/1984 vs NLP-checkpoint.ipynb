# Textual Data Exploration

```python
# Load Van Halen 1984 album
album_folder = "./Data/1984 album Van Halen"
songs = os.listdir(album_folder)
songs
```

```python
# read text in each song file 
album_corpus = []
for title in songs: 
    with open(album_folder + '/' + title, encoding = 'utf8') as f_input:
        album_corpus.append(f_input.read())

album_corpus
    
```

```python
album_word_list = []
for lyrics in album_corpus: 
    for word in lyrics.split(): 
        album_word_list.append(word)
album_word_list
```

```python
len(album_word_list)
```

```python
album_word_count_unique = Counter(album_word_list)
len(album_word_count_unique)
```

```python
# George Orwell 1984 novel
novel = './Data/1984-Novel-George Orwell.txt'
```

```python
novel_corpus = []
with open(novel, encoding='utf8') as f_input:
    novel_corpus.append(f_input.read())
```

```python
novel
novel_corpus[0] = novel_corpus[0].lower()
```

```python
# novel
novel_word_list = novel_corpus[0].split()
```

```python
len(novel_word_list)
```

```python

len(Counter(novel_word_list))
```

# Word Cloud

```python
stop_words = list(get_stop_words("english"))
stop_words.append("[Chorus]")
stop_words.append("[Pre-Chorus]")
stop_words.append("[Spoken]")
stop_words.append("I")
stop_words.append("[Verse")
stop_words.append("Solo]")
```

```python
clean_album_word_list = []
for word in album_word_list: 
    if word not in stop_words: 
        clean_album_word_list.append(word)
len(clean_album_word_list)
```

```python
clean_album_word_count = Counter(clean_album_word_list)
clean_album_word_count
```

```python
import nltk
clean_album_word_count = nltk.FreqDist(clean_album_word_list)
clean_album_word_count
```

```python
clean_album_word_count.plot(50, cumulative=False) 
```

```python
clean_album_word_count_list = Counter(clean_album_word_list)
clean_album_word_count_list
```

```python
wordcloud_1984_album = WordCloud(width= 1000, height=500).generate_from_frequencies(clean_album_word_count_list)
plt.figure(figsize=(20,10))
plt.figtext(0.5,0.9, "WordCloud of 1984 album", fontsize=60, ha = "center")
plt.imshow(wordcloud_1984_album)
plt.axis("off")
```

# novel word Cloud


```python
clean_novel_word_list = []
stop_words = list(get_stop_words("english"))
for word in novel_word_list: 
    if word not in stop_words: 
        clean_novel_word_list.append(word)
len(clean_novel_word_list)
```

```python
frequency_clean_novel_word_list = nltk.FreqDist(clean_novel_word_list)
frequency_clean_novel_word_list.plot(50, cumulative=False) 
```

```python
word_count_clean_novel_corpus = Counter(clean_novel_word_list)
word_count_clean_novel_corpus 
```

```python
word_cloud_1984_novel = WordCloud(width= 1000, height=500).generate_from_frequencies(word_count_clean_novel_corpus)
plt.figure(figsize=(20,10))
plt.figtext(0.5,0.9, "WordCloud of 1984 novel", fontsize=60, ha = "center")
plt.imshow(word_cloud_1984_novel)
plt.axis("off")
```

# NLP Summarization

```python
nltk.download("punkt")
```

```python
#summarise album 

lsa = LsaSummarizer(Stemmer("english")) # lsa = latent semantic analysis
lsa.stopwords = get_stop_words("english")
parser = PlaintextParser.from_string(' '.join(album_corpus), Tokenizer("english"))
album_lsa_summary = lsa(parser.document,10)
album_lsa_summary_list = [str(sentence) for sentence in album_lsa_summary]
album_lsa_summary_list
```

```python
summary_album = ' '.join(album_lsa_summary_list)
print('summary of 1984 album')
print(summary_album)
```

```python
#summarise novel 

lsa = LsaSummarizer(Stemmer("english")) # lsa = latent semantic analysis
lsa.stopwords = get_stop_words("english")
parser = PlaintextParser.from_string(' '.join(novel_corpus), Tokenizer("english"))

```

```python
novel_lsa_summary = lsa(parser.document,3)
novel_lsa_summary_list = [str(sentence) for sentence in novel_lsa_summary]
novel_lsa_summary_list
```

```python
summary_novel = ' '.join(novel_lsa_summary_list)
print('summary of 1984 novel')
print(summary_novel)
```
